```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
sns.set(rc = {'figure.figsize': (5, 5)})
# sns.set_style('whitegrid')
```

```python
data = pd.read_excel("fake job postings filter.xlsx", "Sheet1")
```

# **Pie Chart of Fraudulent Data**

```python
no_fraudulent = data[data['fraudulent'] == 0]

fraudulent = data[data['fraudulent'] == 1]

percentage_no_fraudulent = len(no_fraudulent) / len(data) * 100

percentage_fraudulent = len(fraudulent) / len(data) * 100

fraudulent_pie = np.array([len(no_fraudulent), len(fraudulent)])
fraudulent_labels = ["No Fraudulent\n"+str(len(no_fraudulent))+"\n"+str(round(percentage_no_fraudulent, 2))+"%", "Fraudulent\n"+str(len(fraudulent))+"\n"+str(round(percentage_fraudulent, 2))+"%"]
plt.title("Pie Chart of Fraudulent Data")
plt.pie(fraudulent_pie, labels = fraudulent_labels)
plt.show()
```

This is the pie chart of fraudulent data. It is showed that 17014 jobs are not fraudulent and 866 jobs are fraudulent.
